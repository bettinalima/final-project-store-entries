```python
tourists = pd.read_csv("tourists_per_month_second_clean.csv")
tourists.head()
```

```python
tourists.dtypes
```

```python
tourists = tourists.astype({"Period":str})
```

```python
def period_split_month(row):
    period = row["Period"].split()
    return period[0]
    
```

```python
def period_split_year(row):
    period = row["Period"].split()
    return period[1]
    
```

```python
tourists["Mes"] = tourists.apply(period_split_month,axis=1)
```

```python
tourists["Ano"] = tourists.apply(period_split_year,axis=1)
```

```python
tourists.head()
```

```python
def sub_month_num(row):
    if row == "December":
        return 12
    if row == "November":
        return 11
    if row == "October":
        return 10
    if row == "September":
        return 9
    if row == "August":
        return 8
    if row == "July":
        return 7
    if row == "June":
        return 6
    if row == "May":
        return 5
    if row == "April":
        return 4
    if row == "March":
        return 3
    if row == "February":
        return 2
    if row == "January":
        return 1
```

```python
tourists["Mes"] = tourists["Mes"].apply(sub_month_num)
```

```python
tourists.head()
```

```python
tourists = tourists.drop("Period",axis=1)
```

```python
tourists.head()
```

```python
tourists.dtypes
```

```python
tourists = tourists.astype({"Ano":int})
```
