```python
def season(row):
    if row["Mes"] < 3:
        return "Winter"
    if row["Mes"] == 3 and row["Dia"] < 20:
        return "Winter"
    
    if row["Mes"] == 3 and row["Dia"] >= 20:
        return "Spring"
    if row["Mes"] > 3 and row["Mes"] < 6:
        return "Spring"
    if row["Mes"] == 6 and row["Dia"] < 21:
        return "Spring"
    
    if row["Mes"] == 6 and row["Dia"] >= 21:
        return "Summer"
    if row["Mes"] > 6 and row["Mes"] < 9:
        return "Summer"
    if row["Mes"] == 9 and row["Dia"] < 22:
        return "Summer"
    
    if row["Mes"] == 9 and row["Dia"] >= 22:
        return "Autumn"
    if row["Mes"] > 9 and row["Mes"] < 12:
        return "Autumn"
    if row["Mes"] == 12 and row["Dia"] < 21:
        return "Autumn"
    
    if row["Mes"] == 12 and row["Dia"] >= 21:
        return "Winter"
```

```python
new_data["Season"] = new_data.apply(season,axis=1)
```
